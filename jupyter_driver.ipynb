```python
import sys
from marbl_diags import analysis_elements_class
import yaml
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('sample_input_files/three-level.nitrate.yml') as file_in:
    config_dicts = yaml.load(file_in)
with open('variables.yml') as file_in:
    var_dict = yaml.load(file_in)
# Update configuration dictionary: do not generate png files and
# also save the matplotlib figs as an object on AnalysisElements
config_dicts['CESM2.1_PI-WOA2013-ann-comparison']['plot_format']=None
config_dicts['CESM2.1_PI-WOA2013-ann-comparison']['keep_figs']=True
AnalysisElements = dict()
for config_key, config_dict in config_dicts.items():
    AnalysisElements[config_key] = analysis_elements_class.AnalysisElements(config_key, config_dict, var_dict)
```

```python
%%time
for AnalysisElement in AnalysisElements.values():
    AnalysisElement.do_analysis()
```
